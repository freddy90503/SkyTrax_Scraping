```python
import re
import pandas as pd
import numpy as np
from plotnine import *
from pandas import DataFrame, read_csv
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readData():
    colnames_Top10 = ['AirName','AircraftModel','Comments','DateFlown','EntertainmentRating','FoodRating','GroundServiceRating','OverallScore','Recommended','ReviewDate','ReviewTitle','ReviewrCountry','Route','SeatComfortRating','SeatType','ServiceRating','TravelType','ValueRating','WifiRating']
    ANA = pd.read_csv("./data/ANA.csv",header = None,names = colnames_Top10)
    Asiana = pd.read_csv("./data/Asiana.csv",header = None,names = colnames_Top10)
    Cathay = pd.read_csv("./data/Cathay.csv",header = None,names = colnames_Top10)
    EVA = pd.read_csv("./data/EVA.csv",header = None,names = colnames_Top10)
    Garuda = pd.read_csv("./data/Garuda.csv",header = None,names = colnames_Top10)
    Hainan = pd.read_csv("./data/Hainan.csv",header = None,names = colnames_Top10)
    JapanAir = pd.read_csv("./data/JapanAir.csv",header = None,names = colnames_Top10)
    Lufthansa = pd.read_csv("./data/Lufthansa.csv",header = None,names = colnames_Top10)
    Qatar = pd.read_csv("./data/Qatar.csv",header = None,names = colnames_Top10)
    Singapore = pd.read_csv("./data/Singapore.csv",header = None,names = colnames_Top10)
    df = pd.concat([ANA,Asiana,Cathay,EVA,Garuda,Hainan,JapanAir,Lufthansa,Qatar,Singapore])
    return df.drop('Comments', 1)
```

```python
df_top10 = readData()
```

```python
df_top10 = df_top10.drop_duplicates(keep='first')
```

```python
df_top10 = df_top10.drop([0])
```

```python
def stripChar(x):
    try:
        return re.sub("[()\"]","",x)
    except:
        return np.nan
```

```python
df_top10.ReviewrCountry = df_top10.ReviewrCountry.apply(lambda x: stripChar(x))
```

```python
df_top10.ReviewTitle = df_top10.ReviewTitle.apply(lambda x: stripChar(x))
```

```python
df_top10.Recommended = df_top10.Recommended.fillna(value = 0)
```

```python
df_top10.Recommended = df_top10.Recommended.eq('yes').mul(1)
```

```python
df_top10['ReviewDate'] = pd.to_datetime(df_top10['ReviewDate'])
df_top10['year'] = df_top10['ReviewDate'].dt.year
df_top10['month'] = df_top10['ReviewDate'].dt.month
df_top10
```

```python
Seat = df_top10.groupby('TravelType')['OverallScore'].mean().reset_index().sort_values(by = 'OverallScore', ascending = False)
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(x="TravelType", y="OverallScore", data=Seat, ax=ax)
# ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
for p in ax.patches:
    ax.annotate("%0.2f"%(p.get_height()), (p.get_x(), p.get_height() * 1.005)) 
plt.show()  


```

```python
Seat = df_top10.groupby('SeatType')['OverallScore'].mean().reset_index().sort_values(by = 'OverallScore', ascending = False)
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(x="SeatType", y="OverallScore", data=Seat, ax=ax)
# ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
for p in ax.patches:
    ax.annotate("%0.2f"%(p.get_height()), (p.get_x(), p.get_height() * 1.005)) 
plt.show()                 
                 
                 
                 
```

```python
dff = df_top10.groupby(['AirName','AircraftModel'])['OverallScore'].mean().reset_index()#.sort_values(by = 'OverallScore', ascending = False)
dff1 = df_top10.groupby(['AirName','AircraftModel'])['OverallScore'].count().reset_index()
xx = pd.merge(dff,dff1, on = ['AirName','AircraftModel'])
xx = xx[xx['OverallScore_y'] >=25].sort_values(by = 'OverallScore_x', ascending = False)
xx['Model'] = xx['AirName'] + [' '] +xx['AircraftModel']
xx = xx.head(15)
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(x="Model", y="OverallScore_x", data=xx, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
for p in ax.patches:
    ax.annotate("%0.2f"%(p.get_height()), (p.get_x(), p.get_height() * 1.005)) 
plt.show()
```

```python
test = df_top10
```

```python
test['OverallScore'] = test['OverallScore'].fillna('0')
test['FoodRating'] = test['FoodRating'].fillna('0')
test['EntertainmentRating'] = test['EntertainmentRating'].fillna('0')
test['GroundServiceRating'] = test['GroundServiceRating'].fillna('0')
test['SeatComfortRating'] = test['SeatComfortRating'].fillna('0')
test['ServiceRating'] = test['ServiceRating'].fillna('0')
test['ValueRating'] = test['ValueRating'].fillna('0')
test['WifiRating'] = test['WifiRating'].fillna('0')
test['OverallScore'] = test['OverallScore'].astype('int64')
test['FoodRating'] = test['FoodRating'].astype('int64')
test['EntertainmentRating'] = test['EntertainmentRating'].astype('int64')
test['GroundServiceRating'] = test['GroundServiceRating'].astype('int64')
test['SeatComfortRating'] = test['SeatComfortRating'].astype('int64')
test['ServiceRating'] = test['ServiceRating'].astype('int64')
test['ValueRating'] = test['ValueRating'].astype('int64')
test['WifiRating'] = test['WifiRating'].astype('int64')
```

```python
corrs = test[test != 0].corr()["OverallScore"]
corrs = corrs[~corrs.index.isin(["OverallScore","year","month","Recommended"])].sort_values(ascending=False)
```

```python
stopANA = ['All','Nippon','Airways']
df_ANA = df_top10[df_top10['AirName']=='ANA All Nippon Airways']
df_ANA.ReviewTitle = df_ANA.ReviewTitle.apply(lambda x:" ".join(x for x in x.split() if x not in stopANA))
wc = WordCloud(background_color="white", max_words=300, width=1000, height=600)
# generate word cloud
wc.generate(' '.join(df_ANA['ReviewTitle']))
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
